```python
import sys
sys.path.append("/mnt/lustre/jkyang/PSG4D/hoi4d/HOI4D-Instructions/prepare_4Dseg")
from utils import category2label_map, category2label_map_instanceseg
from palette import pal_color_map
import cv2
import numpy as np
import os


def get_mask_and_label(path):
    l = path.split('/')
    for i in l:
        if 'C' in i:
            category = i
    #for img in os.listdir(path):
        #f(img, category)
    return f(path, category)


def f(img_path, category):
    image = cv2.imread(img_path)[:, :, ::-1]
    assert image.shape == (1080, 1920, 3)
    # image = shift_mask(image, img_path)
    color_map = pal_color_map() #palette

    arrs = []
    labels = []  # semantic segmentation
    labels_instanceseg = []  # instance segmentation
    for i in range(10):
        color = color_map[i + 1]
        valid = (image[..., 0] == color[0]) & (image[..., 1] == color[1]) & (image[..., 2] == color[2])
        if np.sum(valid) == 0:
            continue
        if i >= len(category2label_map[category]): #大于类数肯定不对
            continue
        arrs.append(valid)
        labels.append(category2label_map[category][i])
        labels_instanceseg.append(category2label_map_instanceseg[category][i])
    
    return arrs, np.array(labels), np.array(labels_instanceseg)
```

```python
def getindex(list,catelist):#输入一个list 返回一个字典 包括这个key 他的类
    dict={}
    for i in range(len(list)):
        if(list[i] not in dict.keys()):
            dict[list[i]]=[[i],categories[str(catelist[i])]]#记录他的index 且相同的label的category都相同，就第一次记录就可以了
        else:
            dict[list[i]][0].append(i)
    return dict
```

```python
import warnings
warnings.filterwarnings("ignore", category=DeprecationWarning)
datalist={}
font = ImageFont.truetype("/mnt/lustre/jkyang/PSG4D/liushuai_workdir/OpenSans-Bold.ttf",40)
tickfont = ImageFont.truetype("/mnt/lustre/jkyang/PSG4D/liushuai_workdir/OpenSans-Bold.ttf",40)
for dirpath, dirnames, filenames in os.walk("/mnt/lustre/jkyang/PSG4D/hoi4d/HOI4D_annotations/"):
    if 'mask' in dirpath:
        a=dirpath.split('/')
        videoname=a[-8]+a[-7]+a[-6]+a[-5]+a[-4]+a[-3] #video的命名
        print(videoname)
        out = cv2.VideoWriter(f'/mnt/lustre/jkyang/PSG4D/liushuai_workdir/Hoi4d/hoivideo/{videoname}.mp4', fourcc, 15, (1920,1080))
        instance=a[-7]
        resultdict=getindex(category2label_map_instanceseg[instance],category2label_map[instance])
        num_palette=len(resultdict.keys())
        sub_palette = random.sample(range(1, 256), num_palette)
        for i in range(len(list(resultdict.keys()))):
            resultdict[list(resultdict.keys())[i]].append(palette[sub_palette[i]])
#         print(resultdict)
        videopath='/mnt/lustre/jkyang/PSG4D/hoi4d/HOI4D_release'
        for ele in range(7,len(dirpath.split("/"))-2):
            videopath=os.path.join(videopath,dirpath.split("/")[ele])
        videopath=os.path.join(videopath,"align_rgb","image.mp4")
        cap = cv2.VideoCapture(videopath)
        # 定义要拆分的帧数
        num_frames = 300
        # 确定帧率
        fps = cap.get(cv2.CAP_PROP_FPS)
        # 确定视频尺寸
        frame_size = (int(cap.get(cv2.CAP_PROP_FRAME_WIDTH)), int(cap.get(cv2.CAP_PROP_FRAME_HEIGHT)))
        interval = int(cap.get(cv2.CAP_PROP_FRAME_COUNT) / num_frames)
#         print(videopath)
        tick=0 #计数，记录视频的帧数
        for ele in tqdm.tqdm(sorted(filenames)):
            cap.set(cv2.CAP_PROP_POS_FRAMES, tick*interval)
            _,frame = cap.read() #读视频
            targetdir=os.path.join(dirpath,ele)
#             cap = cv2.VideoCapture()
            mask=get_mask_and_label(targetdir)
            masklist=mask[0]
            finalmasklist=[]
            catelist=mask[1]
            instancelist=mask[2]
            sub_resultdict=getindex(instancelist,catelist) #子字典
            for key in sub_resultdict.keys():
                sub_resultdict[key].append(resultdict[key][2]) #添加进颜色信息
#             print(sub_resultdict)

            foreground=np.zeros(shape=(1080,1920,3)) #最终输出的mask
            for index in sub_resultdict.keys():
                tmpmask=np.zeros(shape=masklist[0].shape)
                for elekey in sub_resultdict[index][0]:#对于每个key的索引，合并他们的mask
                    tmpmask+=masklist[elekey]#tmpmask是一个instance的mask
                rgbmask=np.repeat(tmpmask[:, :, np.newaxis],3,axis=2).astype(np.uint8)
                rgbmask[:,:,0]*=sub_resultdict[index][-1][0]
                rgbmask[:,:,1]*=sub_resultdict[index][-1][1]
                rgbmask[:,:,2]*=sub_resultdict[index][-1][2]
                #背景

                foreground+=rgbmask

            foreground[foreground.sum(axis=2) == 0] = [255, 255, 255]#三层和为0，设置成255
            foreground=foreground.astype('uint8')
            showimg=Image.fromarray(foreground)
            draw = ImageDraw.Draw(showimg)
            cycxlist=[]
            for index in sub_resultdict.keys():
                cy, cx = ndimage.center_of_mass(masklist[sub_resultdict[index][0][0]]==1)
                cycxlist.append((cy,cx))
                text_width, text_height = draw.textsize (f'{sub_resultdict[index][1]}_{index}', font=font)
                draw.rectangle([cx, cy,cx+text_width, cy+text_height],fill=(0,0,0))
                draw.text((cx, cy), f'{sub_resultdict[index][1]}_{index}', fill=tuple(sub_resultdict[index][-1]), font=font)
            text_width, text_height = draw.textsize(f"{ele.strip('.png')[-4:]}", font=tickfont)
            draw.rectangle([0, 0,text_width, text_height],fill=(0,0,0))
            draw.text((0, 0), f"{ele.strip('.png')[-4:]}", fill=(255,255,255), font=tickfont)
            
            alpha=0.4
            expanded_mask=np.array(showimg)
            combine_foreground=(frame * alpha + np.ones(frame.shape) * (1 - alpha) * expanded_mask)
            combine_foreground = combine_foreground.astype('uint8')
            out.write(combine_foreground)
            tick+=1
#             break
        cap.release()
        out.release()
```

```python

# Define projection function
def project(pcd, intrinsic, extrinsic):
    pcd = np.hstack((np.asarray(pcd.points), np.ones((len(pcd.points), 1)))) # add homogeneous coordinates
    pcd = np.dot(extrinsic, pcd.T).T # apply extrinsic matrix
    pcd = np.dot(intrinsic, pcd.T).T[:,:-1] # apply intrinsic matrix 去除齐次坐标
    return o3d.geometry.PointCloud(o3d.utility.Vector3dVector(pcd))
```
